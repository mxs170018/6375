```python
class NeuralNetworkModel:
    def __init__(self, alpha=1.0,optimizer='None'):
        self.alpha = alpha
        self.optimizer = optimizer
        self.numLayers = 0
        self.allLayers = []
    
    # layers = [[5,'Sigmoid',16]]
    # layets = [[units, activation, shape]]
    def sequential(self,layers):
        for layer in layers:
            units,activation,shape = layer
            Wxh = np.random.rand(shape*units).reshape(shape,units) -0.5
            val = []
            self.allLayers.append([activation,Wxh,val])
                
                #print(Wxh)
    def sigmoid(self,num):
        return 1/(1+math.exp(-num))
    
    def tanh(self,num):
        return (math.exp(num) - math.exp(-num))/(math.exp(num)+math.exp(-num))
    
    def relu(self,num):
        if num>0: return num
        else: return 0

    def tanhPrime(self,num):
        return (math.exp(num)**2 - math.exp(-num)**2)/(math.exp(num)**2+math.exp(-num)**2)
    def activate(self,act,arr):
        #print("STUFF",arr)
        if(act == 'Sigmoid'):
            return np.array([self.sigmoid(nums) for nums in arr])
        elif(act == 'Tanh'):
            return np.array([self.tanh(nums) for nums in arr])
        elif(act == 'relu'):
            return np.array([self.relu(nums) for nums in arr])

    def activation(self,act,num):
        if(act == "Sigmoid"):
            return self.sigmoid(num)
        elif(act == "Tanh"):
            return self.tanh(num)
        elif(act == 'relu'):
            return self.relu(num)
    
    def reluPrime(self,num):
        if num>0: return 1 
        else: return 0
    def getOutError(self,act,sigsum,actual):
        if(act == 'Sigmoid'):
            return sigsum*(1-sigsum)*(actual-sigsum)
        elif(act == 'Tanh'):
            return (1-sigsum**2)*(actual-sigsum)
        elif(act == 'relu'):
            return self.reluPrime(sigsum)*(actual-sigsum)
        
    def getHiddenError(self,act,pred,weightDotDelta):
        if(act == 'Sigmoid'):
            return pred*(1-pred)*weightDotDelta
        if(act == 'Tanh'):
            return(1-pred**2)*weightDotDelta
        if(act == 'relu'):
            return self.reluPrime(pred)*weightDotDelta

    def forwardPass(self,layer,data):
        summation = np.dot(data,layer[1])
        out = self.activate(layer[0],summation)
        return out

    def backwardsPass(self):
        return "LOL"
    
            
                
    def recBackPass(self, index, prevX,Y):
        layer = self.allLayers[index]
        errorVector = []
        if(index == (len(self.allLayers)-1)):

            for o_i in range(0,layer[1].shape[1]):
                OutError = self.getOutError(layer[0],layer[2][o_i],Y[o_i])
                errorVector.append(OutError)

            
            
        else:
  
            weightDotDelta = self.recBackPass(index+1,layer[2],Y)
            for o_i in range(0,layer[1].shape[1]):
                
                error = self.getHiddenError(layer[0],layer[2][o_i],weightDotDelta[o_i])
                errorVector.append(error)

        weightDotDelta = np.dot(layer[1],np.transpose(errorVector))

        newOutWeights = []

        for i in range(0,len(layer[1])):
            weightVector = []
            for j in range(0,len(layer[1][i])):
                
                newWeight = layer[1][i][j] + self.alpha*errorVector[j]*prevX[i]
                weightVector.append(newWeight)
            newOutWeights.append(weightVector)
        
        self.allLayers[index][1] = np.array(newOutWeights)
        
        return weightDotDelta            
               
    
    def evaluate(self,testX,testY):

        for row in range(0,testX.shape[0]):
            out = testX[row]
            for layer in self.allLayers:
                out = self.forwardPass(layer,out)
                    
                layer[2] = out

            maxi = out.max()
            yaxi = testY[row].max()
            print("Y",np.where(testY[row]==yaxi)[0],"PRED",np.where(out==maxi)[0])
    def fit(self,trainX,trainY,iterations):
        for i in range(0,iterations):

            for row in range(0,trainX.shape[0]):#trainX.shape[0]
                
                
                out = trainX[row]
                #forward pass
                for layer in self.allLayers:
                    out = self.forwardPass(layer,out)
                    
                    layer[2] = out

                self.recBackPass(0,trainX[row],trainY[row])
        


```
